```python
%matplotlib inline
import fipy as fp
data = fp.dump.read("1bpickles/1b100.pkl")
var = data['var']
# viewer = fp.Viewer(var,
#                   datamin=.2, datamax =.8,
#                   title = "Elapsed Time = 0")
for i in range(100, 1001, 100):
    data = fp.dump.read("1bpickles/1b{0}.pkl".format(i))
    var[:] = data['var']
    viewer = fp.Viewer(var,
                  datamin=.2, datamax =.8,
                  title = "Elapsed Time = {0}".format(i))
    viewer.plot()
```
